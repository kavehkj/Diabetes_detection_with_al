IMPORTING LIBRARY

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import  ConfusionMatrixDisplay,accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

from sklearn.svm import SVC
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

READING DATASET

```python
df = pd.read_csv('human_vital_signs_dataset_2024.csv') 
```

PREPROCESSING AND ENCODING AND SPLIT THE DATASET

```python
le = LabelEncoder()

df['Gender'] = le.fit_transform(df['Gender'])
df['Risk Category'] = le.fit_transform(df['Risk Category'])

df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df['Hour'] = df['Timestamp'].dt.hour


columns_to_drop = ['Patient ID', 'Derived_HRV', 'Derived_Pulse_Pressure', 'Derived_MAP', 'Timestamp']
df.drop(columns=columns_to_drop, axis=1, inplace=True)


X = df.drop('Risk Category', axis=1)
y = df['Risk Category']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

NORMALIZATION

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

DEF FOR REPORT ABOUT MODELS

```python
def evaluate_model(model, X_train, y_train, X_test, y_test, model_name="Model"):

    start_time = time.time()
    
    model.fit(X_train, y_train)

    y_pred_train = model.predict(X_train)
    y_pred_test = model.predict(X_test)

    end_time = time.time()
    elapsed_time = end_time - start_time

    train_acc = accuracy_score(y_train, y_pred_train)
    test_acc = accuracy_score(y_test, y_pred_test)

    print(f"üîπ {model_name} Accuracy on Test Set: {test_acc:.2f}")
    print(classification_report(y_test, y_pred_test))

    print(f"‚úÖ Train Accuracy: {train_acc:.2f}")
    print(f"‚úÖ Test Accuracy: {test_acc:.2f}")
    print(f"‚è±Ô∏è Elapsed Time: {elapsed_time:.2f} seconds")

    cm = confusion_matrix(y_test, y_pred_test)
    disp = ConfusionMatrixDisplay(cm, display_labels=['Low Risk', 'High Risk'])
    disp.plot(cmap=plt.cm.Greens)
    plt.title(f"Confusion Matrix - {model_name}")
    plt.show()
```

LOGISTICREGRESSION

```python
logistic_model = LogisticRegression(max_iter=1000, random_state=42)
evaluate_model(logistic_model, X_train, y_train, X_test, y_test, model_name="Logistic Regression")
```

SVM

```python
svm_model = SVC(kernel='rbf', random_state=42)
evaluate_model(svm_model, X_train, y_train, X_test, y_test, model_name="SVM")
```

ANN 

```python
start_time = time.time()

model = Sequential([
    Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

history = model.fit(X_train, y_train, epochs=20, batch_size=32, validation_split=0.2, verbose=1)

end_time = time.time()
elapsed_time = end_time - start_time

y_pred_prob = model.predict(X_test).flatten()
y_pred = (y_pred_prob > 0.5).astype(int)

print("üîπ ANN Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred, target_names=['Low Risk', 'High Risk']))

y_train_pred_prob = model.predict(X_train).flatten()
y_train_pred = (y_train_pred_prob > 0.5).astype(int)
train_acc = accuracy_score(y_train, y_train_pred)

test_acc = accuracy_score(y_test, y_pred)

print(f"‚úÖ Train Accuracy: {train_acc:.2f}")
print(f"‚úÖ Test Accuracy: {test_acc:.2f}")
print(f"‚è±Ô∏è Elapsed Time: {elapsed_time:.2f} seconds")

cm = confusion_matrix(y_test, y_pred)
ConfusionMatrixDisplay(cm, display_labels=['Low Risk', 'High Risk']).plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix - ANN")
plt.show()
```
